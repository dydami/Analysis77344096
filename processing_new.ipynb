```python
#categorizing strings  for "Technology Used" feild.
used={
'Digital Technology':['Digital','Technology', 'Electronic','Internet'],
 'Artifical Intelligence':['Robotics', 'IoT','AI','Blockchain','Augmented','Reality','Data', 'Science'], 
'Entrepreneurship education':['Entrepreneurship'],
'Other complex tech':['TOEFL', '6+1','MCTS', 'Blended'],
'Unkown':['specified']
}
```

```python
#built a function to tokenize text in "Technology Used"
def other(sample):
    sample=str(sample)
    #sample=sample.lower()
    s=word_tokenize(sample)
    #s=sample.split(" ")
    for y in s:
        #print(used[x])
        for x in used.keys():
            if y in used[x]:
                return x
            else:
                continue
    return sample
```

#COSS-VALIDATION TECHNIQUE FOR COMPARATIVE STUDY OF THE 4 MODELS

```python
len(X)
```

```python
SVM_score=cross_val_score(SVC(C=100),X,Y, cv = cv)
```

```python
MUL_score=cross_val_score(MultinomialNB(alpha=1.0),X,Y, cv = cv)
```

```python
DTC_score=cross_val_score(DecisionTreeClassifier(),X,Y, cv = cv)
```

```python
GUA_score=cross_val_score(GaussianNB(),X,Y, cv = cv)
```

```python
#GETTING AVERAGE SCORES OF THE MODELS ON CROSS VALIDATION
```

```python
SVM_average=np.average(SVM_score)
MUL_average=np.average(MUL_score)
DTC_average=np.average(DTC_score)
GUA_average=np.average(GUA_score)
```
